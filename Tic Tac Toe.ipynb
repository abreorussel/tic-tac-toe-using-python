# Tic Tac Toe

```python
# Numeric grid for the player
def numeric_grid(row1,row2,row3):
    num1 = [1,2,3]     
    num2 = [4,5,6]    
    num3 = [7,8,9]      
         
        
    for i in range(0,3):
        if row1[i] == " ":
            continue
        else:
            num1[i] = " "
    
    for i in range(0,3):
        if row2[i] == " ":
            continue
        else:
            num2[i] = " "
        
    for i in range(0,3):
        if row3[i] == " ":
            continue
        else:
            num3[i] = " "
            
    return f"_{num1[0]}_|_{num1[1]}_|_{num1[2]}_\n\t\t_{num2[0]}_|_{num2[1]}_|_{num2[2]}_\n\t\t {num3[0]} | {num3[1]} | {num3[2]} "
    
        
        
```

```python
# to check if last box is left 
def last_box_check(row1,row2,row3):
    if " " in row1:
        row1[row1.index(" ")] = "X"
        result = winner_check(row1,row2,row3)
        return result
        
    elif " " in row2:
        row2[row2.index(" ")] = "X"
        result = winner_check(row1,row2,row3)
        return result
        
    else: 
        row3[row3.index(" ")] = "X"
        result = winner_check(row1,row2,row3)
        return result
    
    
        
```

```python
#Displaying the grid
def print_grid(row1,row2,row3):
    
    return f"_{row1[0]}_|_{row1[1]}_|_{row1[2]}_\n_{row2[0]}_|_{row2[1]}_|_{row2[2]}_\n {row3[0]} | {row3[1]} | {row3[2]} "
    
    
```

```python
# To check whether that particular player has won
def winner_check(row1,row2,row3):
    
    x_check = ["X","X","X"]
    o_check = ["O","O","O"]
    
    
    if [row1[0],row1[1],row1[2]] ==x_check  or   [row1[0],row1[1],row1[2]]== o_check:
        return True
    elif [row2[0],row2[1],row2[2]] ==x_check  or [row2[0],row2[1],row2[2]]== o_check:
        return True
    elif [row3[0],row3[1],row3[2]] == x_check or [row3[0],row3[1],row3[2]] ==o_check:
        return True
    
    
    elif [row1[0],row2[0],row3[0]] == x_check or [row1[0],row2[0],row3[0]] == o_check:
        return True
    elif [row1[1],row2[1],row3[1]] == x_check or [row1[1],row2[1],row3[1]] == o_check:
        return True
    elif [row1[2],row2[2],row3[2]] == x_check or [row1[2],row2[2],row3[2]] ==o_check:
        return True

        
    elif [row1[0],row2[1],row3[2]] == x_check or [row1[0],row2[1],row3[2]] == o_check:
        return True
    elif [row1[2],row2[1],row3[0]] == x_check or [row1[2],row2[1],row3[0]] == o_check:
        return True
        
    else:
        return False
```

```python
# Updating grid with new values
def updateg(row1,row2,row3,val,pos):
    
    if pos in range(1,4):
        
        if row1[pos-1]==" ":
            row1[pos-1]= val

            if winner_check(row1,row2,row3):
                return "Game Over"
            else:
                return "value inserted"

        else:
                #already inserted
            return "value already there"
        
        
            
    elif pos in range(4,7):
        
        if row2[pos-4]==" ":
            row2[pos-4]=val

            if winner_check(row1,row2,row3):
                return "Game Over"
            else:
                return "value inserted"
        else:
                #already inserted
            return "value already there"
        

    elif pos in range(7,10):
        
        if row3[pos-7]==" ":
            row3[pos-7] = val

            if winner_check(row1,row2,row3):
                return "Game Over"
            else:
                return "value inserted"
        else:
                #already inserted
            return "value already there"
        
            

    else:
        return "out of bounds"
    
```

```python
from IPython.display import clear_output
replay = True
while replay:
    row1 = [" "," "," "]
    row2 = [" "," "," "]
    row3 = [" "," "," "]

    winner = False
    draw = False

    print("Enter the position based on the grid below\n")
    print("_1_|_2_|_3_\n_4_|_5_|_6_\n 7 | 8 | 9\n")
    print("Player 1 : X\nPlayer 2 : O\n")

    while not winner:

        # PLAYER 1 
        #v1 = input("Player 1 \nEnter value: ")

        
        print("----------\nPlayer 1\n----------\n")

        if row1.count(" ") + row2.count(" ") + row3.count(" ") == 1:
            res = last_box_check(row1,row2,row3)
            if res == True:
                winner = True
                break
            else:
                draw = True
                break

        while True:
            p1 = int(input("Enter position for X: "))

            player =1
            if p1 in range(1,10):
                check = updateg(row1,row2,row3,"X",p1)
                break
            else:
                print("ERROR: Position is out of bounds\nRe enter the position")
                continue

        if check== "Game Over":
            winner = True
            continue
        elif check == "value inserted":
            grid = print_grid(row1,row2,row3)
            ref_grid = numeric_grid(row1,row2,row3)
            
            print(f"{grid}\n\t\t{ref_grid}")
            


        # PLAYER 2

    
        print("\n----------\nPlayer 2\n----------\n")

        if row1.count(" ") + row2.count(" ") + row3.count(" ") == 1:
            res = last_box_check(row1,row2,row3)
            if res == True:
                winner = True
                break
            else:
                draw = True
                break

        while True:
            p2 = int(input("Enter position for O: "))

            player =2

            if p2 in range(1,10):
                check = updateg(row1,row2,row3,"O",p2)
                break
            else:
                print("ERROR: Position is out of bounds\nRe enter the position")
                continue

        if check== "Game Over":
            winner = True
            continue
        elif check == "value inserted":
            grid = print_grid(row1,row2,row3)
            ref_grid = numeric_grid(row1,row2,row3)
            
            print(f"{grid}\n\t\t{ref_grid}\n")
            

    grid = print_grid(row1,row2,row3)
    print(grid)

    if draw:
        print("\nITS A DRAW !!!!!!!!!!!")
    else:
        print(f"\nCONGRATULATIONS  Player {player}!!!!!!!!!\n")
        
    
    new_game = input("Do you want to play again ? Yes or No\n\n")
    if new_game.lower().startswith("y") == "yes":
        continue
    elif new_game.lower().startswith("n") == "no":
        break

        
print("Thank You for playing ...\nI hope you enjoyed ...\n\nCOME BACK SOON !!!!!!!!!")
```
